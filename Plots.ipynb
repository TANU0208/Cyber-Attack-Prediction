```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from scipy.signal import butter, lfilter, freqz

train1 = pd.read_csv('train_dataset01.csv')
train2 = pd.read_csv('train_dataset02.csv')
test = pd.read_csv('test_dataset.csv')
```

```python
numerics = train2.select_dtypes(include="float64")
means = numerics.mean()

for col in numerics.columns:
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.plot(numerics[col])
    ax.set_ylabel(col)
    
    color = 'tab:red'
    plt.axhline(y=means[col], color=color)
    plt.show()
```
